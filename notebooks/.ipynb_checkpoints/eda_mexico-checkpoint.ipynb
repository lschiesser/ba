```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
pd.options.display.max_columns = None
os.getcwd()
os.chdir("C:/code/Uni/ba/too-big")
data = pd.read_csv('200817COVID19MEXICO.csv',encoding= 'unicode_escape')
data.head()
```

```python
data.columns[2:]
```

```python
detailed_nan = pd.DataFrame(columns = data.columns)
detailed_nan = detailed_nan.append(pd.Series((((data == 99).sum(axis=0)/data.shape[0])*100), name="NaNs in %"))
for val in data.RESULTADO.unique():
    total = data[data["RESULTADO"] == val].shape[0]
    print(total)
    detailed_nan = detailed_nan.append(pd.Series((((data[data["RESULTADO"] == val] == 99).sum(axis=0)/total)*100), name=f"NaNs for RESULTADO = {val} in %"))
```

```python
detailed_nan
```

```python
fig, axes = plt.subplots(len(data.columns) // 6, 6, figsize=(15, 25))
i = 2
for triaxis in axes:
    for axis in triaxis:
        col = data.columns[i]
        try:
            data[col] = pd.to_numeric(data[col])
        except:
            print(col)
        try:
            data.hist(column = col, bins = 100, ax=axis)
        except:
            axis.hist([])
            axis.set_title(col)
        i = i+1
```
